```python
import pandas as pd

excel_file = 'EXTENDED_1.xlsx'
df = pd.read_excel(excel_file)

df=df.loc[:,['ID','Pickup location of yours for bus','Pickup time of yours for bus','When is your class starting time','When is your class ending time']]
df.columns = ['ID', 'pickup_location','pickup_time','class_start_time','class_end_time']
df.set_index('ID',inplace=True)
```

```python

df1 = pd.read_excel('DATASET_1.xlsx')
df2 = pd.read_excel('DATASET_2.xlsx')
df3 = pd.read_excel('DATASET_3.xlsx')
df4 = pd.read_excel('DATASET_4.xlsx')
df5 = pd.read_excel('DATASET_5.xlsx')
df6 = pd.read_excel('DATASET_6.xlsx')

df1 = df1.iloc[1:,0:3]
df1.columns = ['place_name', 'place_latitude','place_long']
df2 = df2.iloc[1:,0:3]
df2.columns = ['place_name', 'place_latitude','place_long']
df3 = df3.iloc[1:,0:3]
df3.columns = ['place_name', 'place_latitude','place_long']
df4 = df4.iloc[1:,0:3]
df4.columns = ['place_name', 'place_latitude','place_long']
df5 = df5.iloc[1:,0:3]
df5.columns = ['place_name', 'place_latitude','place_long']
df6 = df6.iloc[1:,0:3]
df6.columns = ['place_name', 'place_latitude','place_long']
df7=pd.concat([df1,df2,df3,df4,df5,df6])
df7.index.name = 'ID'
df8=pd.merge(df,df7,on='ID',how='left')
```

```python
df8=df8.dropna(axis = 0, how ='any')
df8['start_hour'] = pd.to_datetime(df8['class_start_time'], format='%H:%M:%S').dt.hour
df8['start_min'] = pd.to_datetime(df8['class_start_time'], format='%H:%M:%S').dt.minute
df8['end_hour'] = pd.to_datetime(df8['class_end_time'], format='%H:%M:%S').dt.hour
df8['end_min'] = pd.to_datetime(df8['class_end_time'], format='%H:%M:%S').dt.minute
df8['class_start_timeD']=df8['start_hour'].astype('float64')+(df8['start_min'].astype('float64')/100)
df8['class_end_timeD']=df8['end_hour'].astype('float64')+(df8['end_min'].astype('float64')/100)

del df8['start_hour']
del df8['start_min']
del df8['end_hour']
del df8['end_min']
```

```python
df8.to_excel('D:\Projectdata.xlsx')
```
